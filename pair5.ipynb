Las columnas land_area, armed_forces_size, co2-emissions, gdp, population, urban_population, latitude, longitude, density, cpi son de tipo object pero en realidad deberían ser numéricas. Crea una función que nos permita cambiar el tipo de dato de estas columnas al que le corresponde.

```python
w_full_c = ['land_area', 'armed_forces_size', 'co2-emissions', 'gdp', 'population', 'urban_population', 'latitude', 'longitude', 'density', 'cpi']
```

```python
w_full[w_full_c]
```

```python
w_full.info()
```

```python
def trans_obj_to_foat(string):
    try:
        return(float(string.replace(',', '')))
    except:
        return(float(string))
```

```python
for columna in w_full_c:
    w_full[columna] = w_full[columna].apply(trans_obj_to_foat)
```

```python
def continentes(pais, pais_a_continente):
    return(pais_a_continente.get(pais))
```
